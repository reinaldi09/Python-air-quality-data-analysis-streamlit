## Informasi Dataset

1. Deskripsi Dataset:

  - No: Nomor urut data.

  - NO2: Konsentrasi nitrogen dioksida (µg/m³).
  - CO: Konsentrasi karbon monoksida (µg/m³).
  - O3: Konsentrasi ozon (µg/m³).
  - TEMP: Suhu udara dalam derajat Celcius (°C).
  - PRES: Tekanan atmosfer dalam hPa.
  - DEWP: Suhu titik embun dalam derajat Celcius (°C).
  - RAIN: Curah hujan (mm).
  - wd: Arah angin (misalnya, N, NNW, dll.).
  - WSPM: Kecepatan angin dalam m/s.
  - station: Nama stasiun pengamatan.

3. Tujuan Dataset: Dataset ini dapat digunakan untuk memahami dan menganalisis:

  - Kualitas udara di suatu wilayah pada berbagai waktu.
Hubungan antara kondisi meteorologi (suhu, kecepatan angin, curah hujan) dengan konsentrasi polutan udara.

**Cleaning PM2.5 Outlier**

```python
Q1 = (airquality_df['PM2.5']).quantile(0.25)
Q3 = (airquality_df['PM2.5']).quantile(0.75)
IQR = Q3 - Q1

maximum = Q3 + (1.5*IQR)
minimum = Q1 - (1.5*IQR)

kondisi_lower_than = airquality_df['PM2.5'] < minimum
kondisi_more_than = airquality_df['PM2.5'] > maximum

airquality_df.drop(airquality_df[kondisi_lower_than].index, inplace=True)
airquality_df.drop(airquality_df[kondisi_more_than].index, inplace=True)
```

### Pertanyaan 2:

2. Apakah terdapat hubungan antara konsentrasi polutan NO2 dan SO2 dengan suhu (TEMP) selama setahun?

Korelasi Pearson atau Spearman, serta scatter plot hubungan NO2-SO2 terhadap suhu.

1. Korelasi pearson => mengukur hubungan linear
2. Korelasi Spearman => hubungan monotonik (tidak harus linear)

```python
# Korelasi Pearson
pearson_corr = monthly_airquality_df[['NO2', 'SO2', 'TEMP']].corr(method='pearson')
print("Korelasi Pearson:")
print(pearson_corr)

# Korelasi Spearman
spearman_corr = monthly_airquality_df[['NO2', 'SO2', 'TEMP']].corr(method='spearman')
print("\nKorelasi Spearman:")
print(spearman_corr)
```

Insight:

1.   Positif: Jika suhu naik, konsentrasi polutan juga meningkat.
2.   Negatif: Jika suhu naik, konsentrasi polutan menurun.
3.   0: Tidak ada hubungan linear (Pearson) atau monoton (Spearman).

1. Korelasi Pearson:
- NO₂ dan SO₂: Korelasi positif sedang (0,318) menunjukkan bahwa peningkatan kadar NO₂ cenderung diikuti oleh peningkatan SO₂, tetapi tidak terlalu kuat.
- NO₂ dan TEMP: Korelasi negatif moderat (-0,440) menunjukkan bahwa saat suhu meningkat, konsentrasi NO₂ cenderung menurun.
- SO₂ dan TEMP: Korelasi negatif yang kuat (-0,631) menunjukkan bahwa peningkatan suhu secara signifikan berhubungan dengan penurunan kadar SO₂.

2. Korelasi Spearman:
- NO₂ dan SO₂: Korelasi positif (0,377) sedikit lebih tinggi dibandingkan Pearson, menunjukkan hubungan monotonic yang sedikit lebih kuat.
- NO₂ dan TEMP: Korelasi negatif serupa dengan Pearson (-0,440), menunjukkan hubungan konsisten bahwa suhu tinggi terkait dengan kadar NO₂ yang lebih rendah.
- SO₂ dan TEMP: Korelasi negatif yang sangat kuat (-0,717), lebih tinggi dari Pearson, menunjukkan hubungan nonlinear di mana peningkatan suhu lebih erat terkait dengan penurunan SO₂.

### Pertanyaan 3:

3. Bagaimana hubungan antara kecepatan angin (WSPM) dan polutan (PM2.5, PM10, CO)?


koefisien korelasi antara WSPM dengan PM2.5, PM10, dan CO.

```python
# Menghitung korelasi Pearson antara WSPM dan polutan
correlation_pm25 = monthly_airquality_df['WSPM'].corr(monthly_airquality_df['PM2.5'])
correlation_pm10 = monthly_airquality_df['WSPM'].corr(monthly_airquality_df['PM10'])
correlation_co = monthly_airquality_df['WSPM'].corr(monthly_airquality_df['CO'])

print(f"Korelasi antara WSPM dan PM2.5: {correlation_pm25:.2f}")
print(f"Korelasi antara WSPM dan PM10: {correlation_pm10:.2f}")
print(f"Korelasi antara WSPM dan CO: {correlation_co:.2f}")
```

1. Korelasi antara WSPM dan PM2.5 (-0,37):
    - Korelasi negatif moderat menunjukkan bahwa kecepatan angin yang lebih tinggi cenderung menurunkan konsentrasi PM2.5. Hal ini mungkin terjadi karena angin membantu menyebarkan partikel kecil di udara.
2. Korelasi antara WSPM dan PM10 (-0,08):
    - Korelasi negatif sangat lemah menunjukkan bahwa kecepatan angin tidak memiliki hubungan yang signifikan dengan konsentrasi PM10. Partikel PM10 yang lebih besar mungkin tidak mudah terbawa angin dibandingkan PM2.5.
3. Korelasi antara WSPM dan CO (0,10):
    - Korelasi positif lemah menunjukkan bahwa kecepatan angin memiliki sedikit hubungan positif dengan konsentrasi CO, meskipun efeknya sangat kecil. Mungkin ada faktor lain yang lebih berpengaruh terhadap konsentrasi CO.

#### CLUSTERING PM 2.5

```python
# Buat clustering manual berdasarkan nilai PM2.5
def categorize_pm25(value):
    if value <= 50:
        return 'Low'
    elif 50 < value <= 100:
        return 'Moderate'
    else:
        return 'High'

# Tambahkan kolom kategori
monthly_airquality_df['PM2.5_Category'] = monthly_airquality_df['PM2.5'].apply(categorize_pm25)

# Warna untuk kategori
colors = {'Low': 'green', 'Moderate': 'orange', 'High': 'red'}

# Visualisasi: Scatter plot PM2.5 vs PM10 dengan kategori
plt.figure(figsize=(10, 6))
for category in colors.keys():
    subset = monthly_airquality_df[monthly_airquality_df['PM2.5_Category'] == category]
    plt.scatter(subset['PM2.5'], subset['PM10'], label=category, color=colors[category], alpha=0.6)

plt.title('Scatter Plot of PM2.5 vs PM10 with Categories')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.legend(title='PM2.5 Category')
plt.grid(True)
plt.show()

# Visualisasi: Histogram distribusi PM2.5 berdasarkan kategori
plt.figure(figsize=(10, 6))
sns.histplot(data=monthly_airquality_df, x='PM2.5', hue='PM2.5_Category', palette=colors, kde=True, bins=30, alpha=0.7)
plt.title('Distribution of PM2.5 by Category')
plt.xlabel('PM2.5')
plt.ylabel('Frequency')
plt.show()
```

#### CLUSTERING NO2

Membagi nilai NO2 menjadi 3 kategori:
Rendah: NO2 ≤ 30
Sedang: 30 < NO2 ≤ 80
Tinggi: NO2 > 80

```python
# Menentukan batas clustering secara manual untuk NO2
def cluster_NO2(value):
    if value <= 30:  # Cluster 1: Rendah
        return 'Rendah'
    elif 30 < value <= 80:  # Cluster 2: Sedang
        return 'Sedang'
    else:  # Cluster 3: Tinggi
        return 'Tinggi'

# Menambahkan kolom cluster untuk NO2
monthly_airquality_df['NO2_Cluster'] = monthly_airquality_df['NO2'].apply(cluster_NO2)

# Statistik distribusi cluster
print(monthly_airquality_df['NO2_Cluster'].value_counts())

# Visualisasi menggunakan countplot untuk melihat distribusi cluster
plt.figure(figsize=(8, 6))
sns.countplot(data=monthly_airquality_df, x='NO2_Cluster', order=['Rendah', 'Sedang', 'Tinggi'], palette='coolwarm')
plt.title('Distribusi Clustering Manual NO2', fontsize=14)
plt.xlabel('Cluster NO2', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualisasi distribusi NO2 berdasarkan cluster
plt.figure(figsize=(10, 6))
sns.boxplot(data=monthly_airquality_df, x='NO2_Cluster', y='NO2', order=['Rendah', 'Sedang', 'Tinggi'], palette='coolwarm')
plt.title('Distribusi Nilai NO2 Berdasarkan Cluster', fontsize=14)
plt.xlabel('Cluster NO2', fontsize=12)
plt.ylabel('Konsentrasi NO2', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
